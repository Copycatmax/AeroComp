### Importing Libraries

```python
from scipy import integrate as integrate
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

1. ### Problem Set 6.1 #10

```python
def f(x): 
    return (np.sqrt(np.sin(x)))**-1

a = 0
b = np.pi/4

solution, err = integrate.quad(f, a, b)
print("The integral of 1/sqrt(sin(x)) from 0 to pi/4 is: {:.4f}".format(solution))
```

3. ### Problem Set 6.2 #1

```python
def f(x): 
    return np.log(x)/(x**2 - 2*x + 2)

a = 1
b = np.pi

ans1, err = integrate.fixed_quad(f, a, b, n=2)
ans2, err = integrate.fixed_quad(f, a, b, n=4)

print("The integral of ln(x)/(x^2-2x+2) from 1 to pi is {0:.4f} with 2 nodes and {1:.4f} with 4 nodes.".format(ans1, ans2))
```

4. ### Problem Set 6.2 #11

```python
x = sp.symbols('x')
y = sp.symbols('y')
a = 2
b = 1
a1 = -2
a2 = 2
f = ((x**2)/(a**2))+((y**2)/(b**2)) - 1
df = sp.diff(f,x)
df = sp.lambdify(x,df)
def f(x):
    return 2 * np.sqrt(1+(df(x)**2))
result, error = integrate.fixed_quad(f, a1, a2)
print("The circumference is: {:.4f}".format(result))
```
